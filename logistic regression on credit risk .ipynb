```python
cr1=cr.iloc[: ,1:13] #remove loan id column.1st column 
```

```python
cr1.Gender.replace({"Male":1,"Female":0},inplace=True)
```

```python
cr1.Married.replace({"Yes":1,"No":0},inplace=True)
```

```python
cr1.Education.replace({"Graduate":1,"Not Graduate":0},inplace=True)
```

```python
cr1.Self_Employed.replace({"Yes":1,"No":0},inplace=True)
```

```python
cr1.Property_Area.replace({"Semiurban":1,"Urban":2,"Rural":3},inplace=True)
```

```python
cr1.Loan_Status.replace({"Y":1,"N":0},inplace=True)
```

```python
cr1.Gender=cr1.Gender.astype("category")
```

```python
cr1.Married=cr1.Married.astype("category")
cr1.Education=cr1.Education.astype("category")
cr1.Self_Employed=cr1.Self_Employed.astype("category")
cr1.Property_Area=cr1.Property_Area.astype("category")
cr1.Loan_Status=cr1.Loan_Status.astype("category")
cr1.Credit_History=cr1.Credit_History.astype("category")
```

```python
cr1=cr1.dropna()
```

```python
cr_x=cr1.iloc[:,0:11]
cr_y=cr1.iloc[:,11]
```

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
cr_x_train , cr_x_test, cr_y_train, cr_y_test = train_test_split(cr_x, cr_y, test_size = .2, random_state = 101)
```

```python
print("Size of X - train: ",cr_x_train.shape)
print("Size of Y - train: ",cr_y_train.shape)
print("Size of X - test: ",cr_x_test.shape)
print("Size of Y - test: ",cr_y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression 
logmodel=LogisticRegression()
logmodel.fit(cr_x_train,cr_y_train)
pred=logmodel.predict(cr_x_test)
pred
```

```python
cr1.describe(include="all")
```

```python
cr.head()
```

```python
from sklearn.metrics import confusion_matrix
```

```python
tab1=confusion_matrix(pred,cr_y_test)
tab1
```

```python
acc=sum(tab1.diagonal())/tab1.sum() #calculate the accuracy
acc
```
